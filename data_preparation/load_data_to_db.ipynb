## Patent to IPC Mapping Table

```python
conn = sqlite3.connect('data/patent.db')
cursor = conn.cursor()

cursor.execute("DROP TABLE IF EXISTS patent_ipc")

cursor.execute("""
CREATE TABLE patent_ipc (
    patent_id TEXT,
    ipc_code TEXT,
    YEAR INTEGER
);
""")
```

```python
data_test = joblib.load('data/embeddings/by_year/embeddings_2005.pkl')
```

```python
data_id = {}

for patent in tqdm(data_test.keys()):
    id = patent.split('_')[0]
    data_id[id] = data_test[patent]
```

```python
for year in tqdm(range(1980, 2024), desc="Years"):
    print(f"Processing year: {year}")
    embedding_file = f'data/embeddings/by_year/embeddings_{year}.pkl'
    data = joblib.load(embedding_file)

    data_id = {}
    
    for patent in data.keys():
        id = patent.split('_')[0]
        data_id[id] = data[patent]
        
    for patent_id, items in data_id.items():
        for ipc_code, embedding in items.items():
            insert_query = """
            INSERT INTO patent_ipc (patent_id, ipc_code, year)
            VALUES (?, ?, ?);
            """
            cursor.execute(insert_query, (patent_id, ipc_code, year))

conn.commit()
```

## Add Metadata

```python
conn = sqlite3.connect('data/patent.db')
cur = conn.cursor()
```

```python
df_metadata = pd.read_parquet('data/patent_metadata.parquet')
```

```python
df_metadata.columns
```

```python
# rename columns
renames = {
    "representative.city": "rep_city",
    "representative.country": "rep_country",
    "representative.name": "rep_name",
    "representative.address": "rep_address",  
    "application.number": "app_number",
    "application.filing_date": "app_date",
    "publication.number": "id",
    "publication.date": "pub_date",
    "publication.country": "pub_country",
    "title.en": "title"
}

df_metadata.rename(columns=renames, inplace=True)

# convert pub_date to datetime
df_metadata['pub_date'] = pd.to_datetime(df_metadata['pub_date'], format='%Y%m%d', errors='coerce')
```

```python
df_metadata.to_sql('patent', conn, if_exists='replace', index=False)
```

```python
# drop duplicate rows from patent_ipc table based on patent_id and ipc_code
cur.execute("""
DELETE FROM patent_ipc
WHERE rowid NOT IN (
    SELECT MIN(rowid)
    FROM patent_ipc
    GROUP BY patent_id, ipc_code
);
""")
conn.commit()
```
